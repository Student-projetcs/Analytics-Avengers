```python
import pycountry
```

```python
def get_country_code(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        if country:
            return country.alpha_3
        else:
            return None
    except LookupError:
        return None

df_latest_btsx['CountryCode'] = df['Country'].apply(get_country_code)

df_latest_btsx
```

```python
fig_map = px.choropleth(df_latest_btsx, locations='CountryCode', color='MortalityRate', hover_name='Country',
                    projection='natural earth',
                    title='Child Mortality Rate by Country (Both Sex, Latest Year)',
                    labels={'MortalityRate': 'Mortality Rate',
                            'Sex' : 'Sex'
                        })

fig_map.show()
```

```python
df_region = df[df['IsLatestYear'] == 1].groupby('Region').mean()

df_region.reset_index(inplace=True)

df_region = df_region.sort_values( by = 'MortalityRate', ascending = False)

df_region
```

```python
variables_of_interest = ['MortalityRate'
                        ]

df_region.sort_values(by='MortalityRate', ascending=False)


# Create a box plot
fig_region_latest = px.bar(df_region, x = 'Region', y=variables_of_interest, 
                color_discrete_map={'MortalityRate': 'purple'}, 
                title = "Child Mortality Rate (Under 5) by Region (Latest Year)", 
                category_orders={"Region": df_region["Region"].tolist()})

fig_region_latest.update_layout(plot_bgcolor='white')

# Display the plot
fig_region_latest.show()
```

```python
df_years = df[df['Year'] >= 2000]

df_years
```

```python
df_country_total_mortality = df_years.groupby('Country')['MortalityRate'].sum().reset_index()

# Merge the total mortality rate back to the original DataFrame
df_years = df_years.merge(df_country_total_mortality, on='Country', suffixes=('', '_Total'))

# Calculate 'MortalityRate_Percentage' within each country
df_years['MortalityRate_Percentage'] = (df_years['MortalityRate'] / df_years['MortalityRate_Total']) * 100
df_years['MortalityRate_Percentage'] = df_years['MortalityRate_Percentage'].round(2)


df_years
```

```python
df_years_sex = df_years[df_years['Sex_Code'] != 'BTSX']

df_years_sex
```

```python
df_years_sex_UK = df_years_sex[df_years_sex['Country'].str.contains('United Kingdom', case = True)]

df_years_sex_UK
```

```python
df_years_sex_UK = df_years_sex[df_years_sex['Country'].str.contains('United Kingdom', case = True)]

fig_uk_years = px.line(df_years_sex_UK, x='Year', y='MortalityRate', color='Sex',
              color_discrete_map={'Male': 'Blue', 'Female': 'Pink'},
              title="Child Mortality Rate (Under 5) in the UK since 2000")

fig_uk_years.show()
```

```python
agg_years_region = {
    'MortalityRate': 'mean',  # Calculate the mean for the 'MortalityRate'
    'MortalityRate_Total': 'mean',     # Take the first value for the 'ValueType'
    'MortalityRate_Percentage': 'mean',          # Take the first value for the 'Year'
    # Add more columns as needed
}



df_years_sex_region = df_years_sex.groupby(['Region', 'Sex', 'Year']).agg(agg_years_region).reset_index()
df_years_sex_region
```

```python
df_years_sex_region_pivot = df_years_sex_region.pivot(index=['Region', 'Sex'], columns='Year', values='MortalityRate')

df_years_sex_region_pivot
```

```python

df_years_sex_region_pivot.drop(['MortalityRate_Average_Change_Percentage', 'Percentage_Change', 'Average_Percentage_Change'], axis=1)
```

```python
df_years_sex_region_pivot['Average_Percentage_Change'] = ((df_years_sex_region_pivot[df_years_sex_region_pivot.columns[-1]] - df_years_sex_region_pivot[df_years_sex_region_pivot.columns[0]]) / df_years_sex_region_pivot[df_years_sex_region_pivot.columns[0]]) * 100

df_years_sex_region_pivot
```

```python
df_avg_mortality = df_years_sex_region_pivot[['MortalityRate_Average_Change_Percentage']]

df_avg_mortality = df_avg_mortality.reset_index()
```

```python
fig_avg_mortality = px.bar(
                    df_avg_mortality, x = 'MortalityRate_Average_Change_Percentage', y = 'Region', color = 'Sex', orientation='h',
                    barmode = 'group', title = 'Avg. Child Mortality Rate Change (2002-2021) by Region'
                    
)

fig_avg_mortality.update_xaxes(title_text="Average Mortality Rate Percentage Change (%)")

fig_avg_mortality.show()
```

```python
fig_years_region = px.line(
    df_years_sex_region, x='Year', y='MortalityRate',
    color='Region', line_group='Sex',  
    line_dash='Sex',  
    markers=True,  
    title='Child Mortality Rate (Under 5) by Region and Sex',
    labels={'Year': 'Year', 'MortalityRate': 'Mortality Rate'},
)

fig_years_region.show()
```
